```python
class Chef:
    
    def MakeChicken(self):
        print("Preparing chicken")
        
    def MakeCurry(self):
        print("Preparing curry")
        
    def MakeRoti(self):
        print("Preparing roti")
```

```python
class ChineseChef:
    def MakeChineseDish(self):
        print("Preparing chinese dish")
```

```python
chef1 = Chef()
```

```python
chef1.MakeChicken()
```

```python
chineseChef1 = ChineseChef()
```

```python
chineseChef1.MakeChineseDish()
```

```python
class ChineseChef(Chef):
    def MakeChineseDish(self):
        print("Preparing chinese dish")
```

```python
chineseChef2 = ChineseChef()
```

```python
chineseChef2.MakeChineseDish()
```

```python
chineseChef2.MakeChicken()
```
