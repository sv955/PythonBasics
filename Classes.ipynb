```python
class Student:
    
    def __init__(self, name, gpa):
        self.name = name
        self.gpa = gpa
        
        
    def ShowDetails(self):
        print(self.name,self.gpa)
        
    def ErrorFunction():
        print(self.name,self.gpa)
        
        
```

```python
student1 = Student("Mohan",9.8)
```
